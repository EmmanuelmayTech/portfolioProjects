```python
#IMPORT LIBRARIES

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure 

%matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8) #adjusts the confirmation of the plots we will use
```

```python
#READ IN THE DATA
df=pd.read_csv(r'C:\Users\ADMIN\OneDrive\Desktop\python\movies.csv')
df.head()
```

```python
df= df.dropna()
#the hunt for missing data
for col in df.columns:
    perc_missing= np.mean(df[col].isnull())
    print('{} - {}%'.format(col, perc_missing))
```

```python
#Datatypes
df.dtypes
```

```python
#switching from float to int
df['budget']=df['budget'].astype('int64')
df['votes']=df['votes'].astype('int64')
df['runtime']=df['runtime'].astype('int64')
df['gross']=df['gross'].astype('int64')
df.dtypes
```

```python
#YEAR AND RELEASE DONT ALL MATCH SO WE HAVE TO RESOLVE THAT BY CREATING A NEW COLUMN OUT OF THE RELEASED COLUMN
df['year_correct'] = df['released'].astype(str).str.extract(r'(\d{4})')
df
```

```python
#method 2
''' Function to extract the year from a date string
def extract_year(date_str):
    try:
        # Split the date string by commas and spaces
        parts = date_str.split()
        # Get the year from the last part and remove parentheses
        year = parts[-1].replace('(', '').replace(')', '')
        return year
    except:
        return None

 Apply the extract_year function to the "date" column and create the "year" column
df['year'] = df['date'].apply(extract_year)

 Display the updated DataFrame
print(df) '''



```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross earnings')
plt.xlabel('Gross earning')
plt.ylabel('Budget for film')
plt.show()
```

```python
#not the best of charts but e can still notice a correlation
#lets plot budget vs gross using seaborn

sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws={"color": "black"}, line_kws = {"color": "red"})
```

```python
#this shows a positive correlation, but we do not know just how much
#lets do a simple corr which works on only numerical
df.corr(numeric_only=True, method = 'pearson') #methods: pearson, kendall, spearman
```

```python
#the budget and gross are highly correlated
#lets visualise this
correlation_matrix = df.corr(numeric_only = True, method = 'pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('CORRELATION MATRIX FOR NUMERIC FEATURES')
plt.show
```
