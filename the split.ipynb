```python
y = df['Class']
X = df.drop('Class',axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2 , random_state=3)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
```

```python
knn.fit(X=X_train, y=y_train)
```

```python
predicted = knn.predict(X = X_test)
expected = y_test
```

```python
expected[:20]
```

```python
predicted[:20]
```

```python
#USE A LIST COMPRESSION TO FIND ANY WRONG PREDICTIONS
wrong_pred = [(p, e) for (p, e) in zip(predicted, expected) if p != e]
```

```python
wrong_pred #knn wrongly predicted 1 out of our 34 values
```

```python
print(f'{knn.score(X_test, y_test):.2%}')
```

```python
#K FOLD CROSS EXAMINATION
from sklearn.model_selection import KFold

kfold = KFold(n_splits=5, random_state = 8, shuffle=True)
#n_splits=5 specifies the numbers of folds
#random_state=3 seeds the random number generated for reproducability
#shuffles=True cause the kfold object to randomize the data by shuffling it before splitting it into folds
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(estimator = knn, X=X_test,
                         y=y_test, cv=kfold)
```

```python
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.linear_model import LogisticRegression, SGDClassifier, Perceptron, RidgeClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import log_loss
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

```

```python
estimators = {
    'Decision Tree' : DecisionTreeClassifier(random_state=3),
    'Random Forest' : RandomForestClassifier(random_state=3),
    'Extra Tree' : ExtraTreesClassifier(random_state=3),
    'Gradient Boost' : GradientBoostingClassifier(random_state=3),
    'AdaBoost' : AdaBoostClassifier(random_state=3),
    'Logistic Regression' : LogisticRegression(random_state=3),
    'SGDC' : SGDClassifier(random_state=3),
    'Ridge' : RidgeClassifier(random_state=3)
}
```

```python
model_names = []
accuracies = []
```

```python
for name, model in estimators.items():
    model.fit(X_train, y_train)
    y_pred= model.predict(X_test)
    # Print the model name
    print(f'{name}')
    # Print the accuracy score
    print(f' Accuracy: {accuracy_score(y_test, y_pred):.3f}')
    # Print the confusion matrix
    print(f' Confusion Matrix: \n{confusion_matrix(y_test, y_pred)}')
    # Print the classification report
    print(f' Report: \n{classification_report(y_test, y_pred)}')
    print("*" * 100)
```
